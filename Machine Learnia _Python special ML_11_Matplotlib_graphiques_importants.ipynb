## Graphique de classification avec plt.scatter

```python
# charger le jeu de données Iris à partir de la bibliothèque sklearn  
iris = load_iris()  

# extraire les données (caractéristiques) et les cibles (étiquettes de classe)  
x = iris.data                     # 'x' contient les caractéristiques, sous forme de tableau (exemples, variables)  
y = iris.target                   # 'y' contient les étiquettes de classe correspondantes pour chaque exemple  
names = list(iris.target_names)   # 'names' contient les noms des classes sous forme de liste  

# afficher le nombre d'exemples et de variables dans le tableau 'x'  
print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')  

# afficher le nombre de classes uniques présentes dans 'y'  
print(f'Il y a {np.unique(y).size} classes')
```

```python
plt.scatter(x[:, 0], x[:, 1], c= y, alpha=0.5, s=100)

#alpha=0.5 : la transparence des points.
#alpha=0 est complètement transparent:
#Les points ne seront pas visibles sur le graphique. C'est comme si ces points n'existaient pas.
#alpha=1 est complètement opaque
# Les points apparaîtront avec leur couleur pleine et aucune transparence.
#ici, alpha=0.5 signifie que les points seront semi-transparents, ce qui peut être utile si plusieurs points se superposent

#s=100 : la taille des points dans le nuage de points.
#Une valeur de 100 signifie que les points seront assez grands.
#L'unité est en pixels.

plt.xlabel('longueur du sépale')      #titre 'longueur du sépale' pour l'axe des x
plt.ylabel('largeur du sépale')      #titre 'largeur du sépale' pour l'axe des 
```

```python
#la couleur opaque signifie que des points sont superposés
```

```python
plt.scatter(x[:, 0], x[:, 1], c= y, alpha=0.5, s=x[:, 2]*100)

#s=x[:, 2]*100 : la taille des marqueurs des points dans le nuage de points.
#ici, la taille est déterminée par les valeurs de la 3ème colonne de la matrice x, multipliées par 100.
#(cela permet d'adapter la taille des points en fonction des données)
#En multipliant par 100, cela permet d'augmenter la taille des points pour qu'ils soient mieux visibles

plt.xlabel('longueur du sépale')      #titre 'longueur du sépale' pour l'axe des x
plt.ylabel('largeur du sépale')      #titre 'largeur du sépale' pour l'axe des y
```

**%matplotlib qt et %matplotlib inline** sont des commandes magiques (magic commands) spécifiques à IPython et Jupyter. Elles permettent de définir le "backend" de Matplotlib, c'est-à-dire le mécanisme utilisé pour afficher les graphiques. Les backends déterminent comment et où les graphiques sont affichés, que ce soit dans une fenêtre séparée, dans le navigateur ou directement dans un notebook Jupyter.

- **%matplotlib qt :** cette commande magique définit le backend de Matplotlib pour utiliser Qt5, une bibliothèque d'interface utilisateur graphique. Lorsqu'on utilise le backend Qt5, les graphiques créés avec Matplotlib seront affichés dans des fenêtres séparées et interactives en dehors du notebook ou de la console IPython. Les fenêtres de graphique Qt5 offrent une interaction plus riche, telle que le zoom, le déplacement et la mise à l'échelle, ainsi que la possibilité de sauvegarder les graphiques dans différents formats de fichiers. Pour utiliser ce backend, ont doit avoir PyQt5 ou PySide2 installé.

- **%matplotlib inline :** cette commande magique (utilisée ici **par défaut**) définit le backend de Matplotlib pour afficher les graphiques directement dans le notebook Jupyter ou la console IPython, sous la cellule qui a généré le graphique. Les graphiques sont intégrés en tant qu'images statiques (format PNG par défaut) et ne sont pas interactifs. Cependant, cette option est pratique pour créer des documents auto-contenus, tels que des rapports, des tutoriels ou des présentations, où on souhaite intégrer des graphiques directement dans le texte et le code.

Pour utiliser ces commandes magiques, il suffit de les exécuter dans une cellule de code IPython ou Jupyter avant de créer les graphiques avec Matplotlib.

### Créer un histogramme en 2D

Créer un histogramme en 2D pour visualiser la distribution des données lorsqu'elles suivent 2 variables.

```python
#créer un histogramme en 2D pour visualiser la distribution des données lorsqu'elles suivent 2 variables
#x[:, 0] : prendre toutes les lignes (:) de la 1ère colonne (index 0) de la matrice x
#x[:, 1] : prendre toutes les lignes (:) de la 2ème colonne (index 1) de la matrice x
plt.hist2d(x[:, 0], x[:, 1])

#donner des titres aux axes
plt.xlabel('longueur du sépale')      #titre 'longueur du sépale' pour l'axe des x
plt.ylabel('largeur du sépale')      #titre 'largeur du sépale' pour l'axe des y
```

```python
#créer un histogramme en 2D pour visualiser la distribution des données lorsqu'elles suivent 2 variables
#x[:, 0] : prendre toutes les lignes (:) de la 1ère colonne (index 0) de la matrice x
#x[:, 1] : prendre toutes les lignes (:) de la 2ème colonne (index 1) de la matrice x
plt.hist2d(x[:, 0], x[:, 1])

#donner des titres aux axes
plt.xlabel('longueur du sépale')      #titre 'longueur du sépale' pour l'axe des x
plt.ylabel('largeur du sépale')      #titre 'largeur du sépale' pour l'axe des y

#ajouter la barre de couleurs pour bien comprendre l'hist en 2D
plt.colorbar()
```

### L'utilisation de l'histogramme pour l'analyse d'une image

```python
#importer divers éléments depuis scipy pour accéder aux exemples d'images
from scipy import misc

#charger l'image en couleur gris à partir de scipy
#et l'enregistrer dans un tableau numpy appelé 'face'
face = misc.face(gray=True)

#afficher l'image en utilisant une palette de couleurs grises
plt.imshow(face, cmap='gray')

#afficher lq figure avec bins=255
plt.hist(bins=255)

#bins=255 détermine les valeurs de pixels dans l'image (de 0,pour le noir,à 255, pour le blanc)
```

```python
#importer divers éléments depuis scipy pour accéder aux exemples d'images
from scipy import misc

#charger l'image en couleur gris à partir de scipy
#et l'enregistrer dans un tableau numpy appelé 'face'
face = misc.face(gray=True)

#afficher lq figure avec bins=255
#face.ravel() aplatit le tableau d'images 2D en 1D pour le calcul de l'histogramme
plt.hist(face.ravel(), bins=255)

#qfficher la figure
plt.show()
```

```python
#l'hist montre les nombres d'apparitions des différentes valeurs de pixels (de 0 à 255) dans l'image
```

```python
# Charger l'image en couleur gris à partir de scipy
# Cette fonction charge un exemple d'image de visage en niveaux de gris
face = misc.face(gray=True)

# Création d'une figure pour afficher l'image et son histogramme
plt.figure(figsize=(10, 5)) # Définir la taille de la figure

# Affichage de l'image à l'aide de la première sous-figure
plt.subplot(1, 2, 1) # Créer une grille 1x2 et utiliser le premier sous-graphique
plt.imshow(face, cmap='gray') # Afficher l'image en utilisant une palette de couleurs grises
plt.axis('off') # Désactiver l'axe pour un aspect plus net
plt.title('Image du Visage') # Titre de l'affichage de l'image

# Création d'un histogramme des valeurs de gris de l'image
plt.subplot(1, 2, 2) # Utiliser le deuxième sous-graphique
plt.hist(face.ravel(), bins=255, range=[0, 255], color='black', alpha=0.7)
# face.ravel() aplatit le tableau d'images 2D en 1D pour le calcul de l'histogramme
# bins=255 détermine le nombre de bins pour les valeurs de pixels (0 à 255)
# color='black' définit la couleur des barres de l'histogramme
# alpha=0.7 définit la transparence des barres

plt.title('Histogramme des Valeurs Grises') # Titre de l'histogramme
plt.xlabel('Intensité des Pixels') # Libellé de l'axe des X pour l'intensité des pixels
plt.ylabel('Fréquence') # Étiquette de l'axe Y pour la fréquence

plt.tight_layout() # Ajustez la mise en page pour éviter le chevauchement entre les sous-parcelles
plt.show() # Affichez la figure
```

## Contour plot

Utilisé pour visualiser un modèle en 3D en vue du dessus.

### 1er exemple

```python
#tout à l'heure, on a vu l'exemple:

#créer la fonction lambda, f(x, y)= sin(x) + cos(x+y)
f = lambda x, y: np.sin(x) + np.cos(x+y)

#créer un ensemble de 100 points (vecteur numpy X) allant de 0 à 5
X = np.linspace(0, 5, 100)

#créer un ensemble de 100 points (vecteur numpy Y) allant de 0 à 5
Y = np.linspace(0, 5, 100)

#créer une grille à partir des 2 axes X et Y
X, Y = np.meshgrid(X,Y)

# Z = f(X,Y)
Z = f(X,Y)

#créer un objet 'ax_2' sur lequel on va travailler
ax_2 = plt.axes(projection= '3d')

#tracer le graphique f(X,Y), choisir la couleur 'plasma' pour le graph
ax_2.plot_surface(X, Y, Z, cmap='plasma')
```

```python
#contour plot de Z= f(X,Y), en vue du dessus
plt.contour(X, Y, Z)
```

```python
#contour plot de Z= f(X,Y), en vue du dessus
#avec 20 niveaux, pour augmenter la précision de l'affichage du graph
plt.contour(X, Y, Z, 20)
```

```python
#contour plot de Z= f(X,Y), en vue du dessus
#avec 20 niveaux, pour augmenter la précision de l'affichage du graph
#en couleur noir
plt.contour(X, Y, Z, 20, colors='Black')
```

```python
#les valeurs positives s'affichent avec un trait continu
##les valeurs positives s'affichent avec un trait pointillé
```

### 2ème exemple

```python
#tout à l'heure, on a vu l'exemple:

#créer la fonction lambda, f(x, y)= sin(x) + cos(x+y)*cos(x)
f = lambda x, y: np.sin(x) + np.cos(x+y)*np.cos(x)

#créer un ensemble de 100 points (vecteur numpy X) allant de 0 à 5
X = np.linspace(0, 5, 100)

#créer un ensemble de 100 points (vecteur numpy Y) allant de 0 à 5
Y = np.linspace(0, 5, 100)

#créer une grille à partir des 2 axes X et Y
X, Y = np.meshgrid(X,Y)

# Z = f(X,Y)
Z = f(X,Y)

#créer un objet 'ax_2' sur lequel on va travailler
ax_2 = plt.axes(projection= '3d')

#tracer le graphique f(X,Y), choisir la couleur 'plasma' pour le graph
ax_2.plot_surface(X, Y, Z, cmap='plasma')
```

```python
#contour plot de Z= f(X,Y), en vue du dessus
#avec 20 niveaux, pour augmenter la précision de l'affichage du graph
#en couleur noir
plt.contour(X, Y, Z, 20, colors='Black')
```

```python
#contour plot de Z= f(X,Y)
#chosir la couleur cmap='RdGy'
plt.contourf(X, Y, Z, cmap='RdGy')
```

```python
#contour plot de Z= f(X,Y)
#avec 20 niveaux, pour augmenter la précision de l'affichage du graph
#chosir la couleur cmap='RdGy'
plt.contourf(X, Y, Z, 20, cmap='RdGy')
```

```python
#contour plot de Z= f(X,Y)
#avec 20 niveaux, pour augmenter la précision de l'affichage du graph
#chosir la couleur cmap='RdGy'
plt.contourf(X, Y, Z, 20, cmap='RdGy')

#ajouter une barre de couleurs
plt.colorbar()
```
