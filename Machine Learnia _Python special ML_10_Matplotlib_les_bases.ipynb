## 1er graphique avec plt.plot()

### Pyplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#créer un tableau de 10 points de 0 à 2
x = np.linspace(0, 2, 10)

#y = x^2
y = x**2

#afficher x
x
```

```python
#afficher y
y
```

```python
#tracer y = x^2
plt.plot(x,y)
```

```python
#Si on utilise autre chose que jupyter notebook pour afficher le graph, il faut écrire plt.show():

#plt.plot(x,y)
#plt.show() <<<
```

```python
#tracer y = x^2 en utilisant un nuage de point
plt.scatter(x,y)
```

### Les styles les plus importants

- **label =** nom de la courbe

- **lw (line width) =** épaisseur du trait

- **ls (line style) =** style du trait

- **c (color) =** couleur du trait

#### c (color) = couleur du trait

```python
#tracer y = x^2 en utilisant la couleur rouge
plt.plot(x, y, c='red')
```

```python
#tracer y = x^2 en utilisant la couleur noire
plt.plot(x, y, c='black')
```

#### lw (line width) = épaisseur du trait

```python
#tracer y = x^2 en utilisant la couleur noire et lw=3 (trait plus ou moins épais)
plt.plot(x, y, c='black', lw=3)
```

```python
#tracer y = x^2 en utilisant la couleur noire et lw=5 (trait épais)
plt.plot(x, y, c='black', lw=5)
```

#### ls (line style) = style du trait

```python
#tracer y = x^2 en utilisant la couleur noire, lw=5, et le style de ligne pointillée(--)
plt.plot(x, y, c='black', lw=5, ls='--')
```

```python
#tracer y = x^2 en utilisant la couleur noire, lw=5, et le style de ligne pointillée (:)
plt.plot(x, y, c='black', lw=5, ls=':')
```

### Le cycle de vie d'une figure Matplotlib (IMPORTANT)

```python
#créer une figure et définir sa taille en pouces (pas en cm)
plt.figure(figsize= (12, 8))

#créer ensuite le 1er graphiaue y = f(x), ici y = x^2
plt.plot(x, y)
```

```python
#créer une figure vide
plt.figure()

#créer ensuite les graphiaues y = x^2 et y = x^3
plt.plot(x, y)      #y = x^2
plt.plot(x, x**3)   #y = x^3
```

```python
#créer une figure vide
plt.figure()

#créer ensuite les graphiaues y = x^2 et y = x^3
plt.plot(x, y)      #y = x^2
plt.plot(x, x**3)   #y = x^3

#ajouter un titre au graphique
plt.title('Figure 1')
```

```python
#créer une figure vide
plt.figure()

#créer ensuite les graphiaues y = x^2 et y = x^3
plt.plot(x, y)      #y = x^2
plt.plot(x, x**3)   #y = x^3

#ajouter le titre 'Figure 1' au graphique
plt.title('Figure 1')

#ajouter le titre 'axe x' à l'axe des x
plt.xlabel('axe x')

#ajouter le titre 'axe y' à l'axe des y
plt.ylabel('axe y')
```

```python
#créer une figure vide
plt.figure()

#créer ensuite les graphiaues y = x^2 et y = x^3
plt.plot(x, y, label = 'Quadratique')      #y = x^2, ajouter la légende 'Quadratique'
plt.plot(x, x**3, label = 'Cubique')   #y = x^3, ajouter la légende 'Cubique'

#ajouter le titre 'Figure 1' au graphique
plt.title('Figure 1')

#ajouter le titre 'axe x' à l'axe des x
plt.xlabel('axe x')

#ajouter le titre 'axe y' à l'axe des y
plt.ylabel('axe y')

#ajouter une légende au graphique
plt.legend()

#Regarder les détails ajoutés en haut de ce code
```
