### subplot: afficher plusieurs figures avec Matplotlib

**Subplot** permet de générer une grille de plusieurs graphiques sur une figure.

**plt.subplot(lignes, colonnes, position)**

```python
#créer 2 graphs sur une seule colonne
plt.subplot(2, 1, 1)       #2 lignes, 1 colonne, puis commencer par le graph en position 1
plt.plot(x, y, c= 'red')   #tracer y = f(x) en rouge
```

```python
#créer 2 graphs sur une seule colonne
plt.subplot(2, 1, 1)           #2 lignes, 1 colonne, puis commencer par le graph en position 1
plt.plot(x, y, c= 'red')       #tracer y = f(x) en rouge

plt.subplot(2, 1, 2)           #le graph en position 2 (sans la courbe)
```

```python
#créer 2 graphs sur une seule colonne
plt.subplot(2, 1, 1)           #2 lignes, 1 colonne, puis commencer par le graph en position 1
plt.plot(x, y, c= 'red')       #tracer y = f(x) en rouge

plt.subplot(2, 1, 2)           #le graph en position 2 (sans la courbe)
plt.plot(x, y, c= 'blue')      #tracer y = f(x) en bleu
```

```python
#2 graphiques qui font partie d'une même figure
```

Au final, voilà à quoi ressemble le cycle de vie d'une figure à l'intérieur de Matplotlib:

```python
#créer une figure
plt.figure()

#si on veut un système de grille (2 lignes, 1 colonne), on crée un subplot
plt.subplot(2,1, 1)              # graph en position 1
plt.plot(x, y, label = 'x^2')                   #tracer y = f(x), ici y = x^2, ajouter un label au graph
plt.title('Graphique 1')         #donner le titre 'Graphique 1' à ce graph
plt.legend()                     #ajouter une légende au 1er graph


plt.subplot(2, 1, 2)              # graph en position 2
plt.plot(x, np.sin (x), label = 'sin(x)')           #tracer y = sin(x), ajouter un label au graph
plt.plot(x, np.cos (x), label = 'cos(x)')           #tracer y = cos(x), ajouter un label au graph
plt.title('Graphique 2')         #donner le titre 'Graphique 1' à ce graph
plt.legend()                     #ajouter une légende au 2ème graph
```
